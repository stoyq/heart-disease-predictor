```python
import os
import pandas as pd
import pandas as pd
import matplotlib.pyplot as plt
```

### 1. Load the Dataset and inspect the column

```python
file_path = "../data/water-hydrants.csv"
hydrants = pd.read_csv(file_path, sep=";")

#hydrants.columns
```

### 2. Convert Date Column + Extract Year

```python
# Extract only the year from instalation Date Column (INSTDATE1)
hydrants["Year_Installed"] = pd.to_datetime(hydrants["INSTDATE1"]).dt.year.astype("Int64")

hydrants[["INSTDATE1", "Year_Installed"]].head()
```

### 3. Question 1: Hydrants Installed Per Year

```python
# How many hydrants are installed per year?

install_per_year = (
    hydrants.groupby("Year_Installed")
            .size()
            .reset_index(name="count")
            .sort_values("Year_Installed")
)
install_per_year
```

### 4. Plot: Number of Hydrant Installations Per Year

```python
plt.figure(figsize=(12, 6))
plt.plot(
    install_per_year["Year_Installed"],
    install_per_year["count"],
    marker="o",
    linewidth=2
)

plt.title("Hydrants Installed Per Year", fontsize=14)
plt.xlabel("Year Installed")
plt.ylabel("Number of Hydrants Installed")
plt.grid(True)

plt.show()
```
