### Introduction

The City of Vancouver maintains thousands of water (fire) hydrants across the city. These hydrants vary in installation year, flow-rate (represented by color), and operational status.

### Objective 
The goal of this project is to build a prediction model that recommends new hydrant installation locations across Vancouver.

### Analysis
This dataset contains the geographic coordinates and attributes of all existing hydrants in Vancouver. Using this historical installation data, we can analyze spatial patterns and build a machine-learning model to identify which areas of the city are most likely candidates for future hydrant installation.

### Methodology
We aim to identify the most likely area in Vancouver for installing a new hydrant using a K-Nearest Neighbours (KNN) prediction model built from historical hydrant locations.
